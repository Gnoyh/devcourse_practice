```python
import seaborn as sns
```

```python
sns.lineplot(x=[1, 3, 2, 4], y=[0.7, 0.2, 0.1, 0.05])
```

```python
sns.barplot(x=["Amy", "Bob", "Cat", "Dog"], y=[0.7, 0.2, 0.1, 0.05])
```

```python
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=["Amy", "Bob", "Cat", "Dog"], y=[0.7, 0.2, 0.1, 0.05])
plt.title("Bar Plot")
plt.xlabel("X Label")
plt.ylabel("Y Label")

plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x=[1, 3, 2, 4], y=[0.7, 0.2, 0.1, 0.05])
plt.ylim(0, 2)

plt.show()
```
